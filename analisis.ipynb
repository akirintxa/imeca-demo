## Unir los dataframes, crear columna de Sede

```python
# Unir los dos DataFrames
df_total = pd.concat([df_del, df_pal], ignore_index=True)
df_total.info()
```

```python
# Crea una nueva columna Sede
df_total['Sede'] = df_total['Invoice No'].str[:3]

```

```python
print(df_total.head())
```

```python
#mover la columna sede al comienzo del dataframe
# Primero, extraemos la columna 'Sede' de su posición actual
columna_sede = df_total.pop('Sede')

# Luego, la insertamos en la posición 0 (la primera columna)
df_total.insert(0, 'Sede', columna_sede)
```

```python
df_total.info()
```

## Cálculos globales

```python
#agregar una columna con el total vendido en cada fila si hiciera falta
df_total['Total Sales'] = df_total['Invoice Quantity'] * df_total['Real Sale Price']
```

```python
print (df_total.head())
```

##### Ventas globales

```python
#calcular el total de ventas por sede
ventas_por_sede = df_total.groupby('Sede')['Total'].sum().reset_index()
# Aplicamos formato de moneda
ventas_por_sede['Total'] = ventas_por_sede['Total']
print(ventas_por_sede)
```

```python
# Definimos el gráfico de barras
plt.bar(ventas_por_sede['Sede'], ventas_por_sede['Total'])

# Definimos una lista de colores (tantos como sedes tengas)
mis_colores = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'] # Ejemplo con códigos Hex

# Pasamos la lista de colores al argumento 'color'
plt.bar(ventas_por_sede['Sede'], ventas_por_sede['Total'], color=mis_colores)

# Etiquetas y título
plt.xlabel('Sede')
plt.ylabel('Ventas Totales (US$)')
plt.title('Comparación de Ventas por Sede')

# Opcional: Agregar los valores sobre las barras para mayor claridad
for i, valor in enumerate(ventas_por_sede['Total']):
    plt.text(i, valor + 0.5, f'US${valor:,.0f}', ha='center', fontsize=10)

# Ajustamos el eje Y para que tenga espacio arriba
plt.ylim(0, ventas_por_sede['Total'].max() * 1.2)


# Mostramos el gráfico
plt.show()
```

#### Margen por sede

##### Margen global

```python
# calcular el margen por sede (Real Sale Price - Cost Price)/ Real Sale Price * 100
df_total['Margin %'] = (df_total['Real Sale Price'] - df_total['Unit Cost']) / df_total['Real Sale Price'] * 100
margen_por_sede = df_total.groupby('Sede')['Margin %'].mean().reset_index()
print(margen_por_sede)
```

```python
# 1. Calculamos el margen promedio por sede y redondeamos a 2 decimales
margen_por_sede = df_total.groupby('Sede')['Margin %'].mean().round(2).reset_index()

print("Margen promedio por sede (%):")
print(margen_por_sede)

# 2. Definimos colores para las barras
colores = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']

# 3. Creamos el gráfico
plt.bar(margen_por_sede['Sede'], margen_por_sede['Margin %'], color=colores)

# Etiquetas y título
plt.xlabel('Sede')
plt.ylabel('Margen Promedio (%)')
plt.title('Margen de Utilidad Promedio por Sede')

# Opcional: Agregar los valores sobre las barras para mayor claridad
for i, valor in enumerate(margen_por_sede['Margin %']):
    plt.text(i, valor + 0.5, f'{valor}%', ha='center', fontsize=10)

# Ajustamos el eje Y para que tenga espacio arriba
plt.ylim(0, margen_por_sede['Margin %'].max() * 1.2)

plt.show()
```

##### Margen mensual


```python
# Agrupamos por mes y sede, calculando el margen promedio
margen_mensual = df_total.groupby(['YearMonth', 'Sede'])['Margin %'].mean().reset_index()

# Pivotamos para tener las sedes como columnas (formato ideal para graficar)
margen_pivot = margen_mensual.pivot(index='YearMonth', columns='Sede', values='Margin %')

# Convertimos el índice a string para que el gráfico sea más legible
margen_pivot.index = margen_pivot.index.astype(str)
```

```python
# Aplicamos media móvil para suavizar la tendencia
margen_suave = margen_pivot.rolling(window=2, min_periods=1).mean()

# Graficamos
fig, ax = plt.subplots(figsize=(12, 6))
margen_suave.plot(ax=ax, linewidth=2)

# Añadimos líneas de promedio individual por sede
for sede in margen_pivot.columns:
    valor_promedio = margen_pivot[sede].mean()
    # Buscamos el color asignado a la línea de tendencia
    color_linea = next(line.get_color() for line in ax.get_lines() if line.get_label() == sede)
    
    ax.axhline(y=valor_promedio, color=color_linea, linestyle=':', alpha=0.7, 
               label=f'Promedio {sede}: {valor_promedio:.2f}%')

# Formato
plt.title('Tendencia del Margen Porcentual por Sede (Promedio Mensual)')
plt.ylabel('Margen Promedio (%)')
plt.xlabel('Mes')
plt.grid(alpha=0.3)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
