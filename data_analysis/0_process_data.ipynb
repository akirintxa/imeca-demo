````python
# Carga los archivos de dos sucursales en dos dataframes
df_del = pd.read_csv('../stores_data/raw/ventas_DEL.csv')
df_pal = pd.read_csv('../stores_data/raw/ventas_PAL.csv')
````

## Process and merge data

````python
# Unir los dos DataFrames en uno
df_total = pd.concat([df_del, df_pal], ignore_index=True)
df_total.info()
````

````python
# Crea una nueva columna Sede
df_total['Sede'] = df_total['Invoice No'].str[:3]
print(df_total.head())
````

````python
#mover la columna sede al comienzo del dataframe
# Primero, extraemos la columna 'Sede' de su posición actual
columna_sede = df_total.pop('Sede')

# Luego, la insertamos en la posición 0 (la primera columna)
df_total.insert(0, 'Sede', columna_sede)

df_total.info()
````

## Create a CSV (ventas.csv) file with 2 stores

````python
# Exportar a CSV
df_total.to_csv('../stores_data/processed/ventas.csv', index=False)
````
